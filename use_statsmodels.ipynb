## ARMAモデルに従う時系列を生成するクラス

```python
from statsmodels.tsa.arima_process import arma_generate_sample
np.random.seed(0)
y = arma_generate_sample(ar=[1.0, -0.5, 0.7, -0.4], ma=[1.0, -0.7], nsample=1000, distrvs=NormUniform, burnin=500) 
print(y.shape)
fig = plt.figure(figsize=(16,4))
ax = plt.plot(y)
```

```python
def NormUniform(size = 1000):
    if type(size)!=int:
        size = size[0]
    return np.concatenate([np.random.standard_normal(size=size//2), np.random.rand((size+1)//2)])
```

```python
plt.plot(NormUniform(size = 1000))
```

### 与えられたデータにARMAモデルをフィットさせた時の情報量基準を最小化させるpとqの値

```python
order = sm.tsa.arma_order_select_ic(y=y, max_ar=5, max_ma=3, ic=["aic", "bic"], trend="nc")
print(order.aic_min_order)
print(order.bic_min_order)
```
