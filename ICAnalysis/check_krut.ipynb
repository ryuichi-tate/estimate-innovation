# 分布の尖度を調べる

```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
```

$kurt(x)=E[x^4]-3(E[x^2])^2$

正規分布

```python
x = stats.norm.rvs(loc=0, size=10000)
plt.hist(x=x, bins=100)
plt.show()
(x**4).mean() - 3*((x**2).mean())**2
```

一様分布

```python
x = np.random.uniform(low=-np.sqrt(3), high=np.sqrt(3), size=10000)
plt.hist(x=x, bins=10)
plt.show()
(x**4).mean() - 3*((x**2).mean())**2
```

移動ベータ分布

```python
x = np.random.beta(a=0.5, b=0.5, size=10000)
x = x-x.mean()
plt.hist(x=x, bins=100)
plt.show()
(x**4).mean() - 3*((x**2).mean())**2
```

```python
x = np.random.beta(a=500, b=5, size=10000)
# x = x-x.mean()
plt.hist(x=x, bins=100)
plt.show()
(x**4).mean() - 3*((x**2).mean())**2
```

ラプラス分布

```python
x = np.random.laplace(loc=0, scale=1, size=10000)
plt.hist(x=x, bins=100, density=True)
plt.plot(np.linspace(start=-5, stop=5, num=1000), stats.laplace.pdf(x=np.linspace(start=-5, stop=5, num=1000), loc=0))
plt.show()
(x**4).mean() - 3*((x**2).mean())**2
```
